· Acessar o Colab, carregar os módulos e carregar os dados.

```python
import pandas as pd

challenge_df = pd.read_csv('assets/dados_1997_2011_paises.csv', sep=';', decimal=',', encoding='latin1')

challenge_df
```


· Aplicar filtro do país Brasil e ano = 2008.

```python
brasil = challenge_df[challenge_df['pais'] == 'Brasil']
ano = challenge_df[  challenge_df['ano'] == 2008]

result = pd.merge(brasil, ano)

result
```

```python
# or
challenge_df[(challenge_df['pais'] == 'Brasil') & (challenge_df['ano'] == 2008)]
```

· Utilizando o groupby, obtenha o idh média para cada país.


```python
challenge_df.groupby('pais').describe()
```

```python
challenge_df.groupby('pais')['idh'].mean().reset_index()
```
