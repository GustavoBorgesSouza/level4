# Matplotlib and numpy

```python
%matplotlib inline
# used to display graphics in jupyter notebooks

import matplotlib.pyplot as plt
import numpy as np

arr = np.arange(12)
plt.plot(arr)

plt.show()
```

```python
arr = np.arange((30))

arr2 = [number**2 for number in arr]
plt.plot(arr2)
plt.show()
```

```python
eixoX = np.linspace(0,2,100)
seno = np.sin(2*np.pi*eixoX)

plt.plot(eixoX, seno)

plt.show()
```

```python
eixoX = np.linspace(0,2,20)
seno = np.sin(2*np.pi*eixoX)

plt.title('Grfico super maneiro')
plt.xlabel('Valor do eixo x')
plt.ylabel('Valor do eixo y')
plt.grid(True)
plt.plot(eixoX, seno, color='tab:red', linestyle='-', marker='s'); plt.show();


plt.plot(eixoX, seno, 'm:', drawstyle='steps-post'); plt.show();


x = np.arange(seno.size)
plt.bar(x, seno)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
eixoX=np.linspace(0,2,20) 
seno =np.sin(2*np.pi*eixoX)

#Subplot 1  Cor ou estilo ou marcador
plt.subplot(2,2,1) 
plt.plot(eixoX,seno,'g'); #Apenas cor

#Subplot 2  Cor + estilo
plt.subplot(2,2,2) #Nomear os parmetros  opcional
plt.plot(eixoX, seno, 'k-.')

#Subplot 3  Cor + marcador
plt.subplot(2,2,3)
plt.plot(eixoX,seno,'y*');

#Subplot 4  Cor + estilo + marcador
plt.subplot(2,2,4)
plt.plot(eixoX,seno,'r-o');
```
